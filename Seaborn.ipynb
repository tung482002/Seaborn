```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

tungvip = pd.read_csv('/tungvip.txt')
tungvip.head() 
tip_df = sns.load_dataset('tips')
tip_df.head()

sns.histplot(data = tip_df['total_bill']);
```

```python
tip_df['total_bill'].value_counts().sort_values(ascending = False)
```

```python
sns.kdeplot(data = tip_df['total_bill'])
```

```python
sns.displot(data = tip_df, x = 'total_bill', kde = True, col = 'time')
```

```python
sns.barplot(data = tip_df, x = 'sex', y='tip', estimator=np.mean);
```

```python
sns.boxplot(data = tip_df, x = 'day', y = 'total_bill', hue = 'sex')
plt.legend(loc = 0)
```

```python
tip_fg = sns.FacetGrid(data = tip_df, row = 'smoker', col = 'time')
tip_fg.map(sns.scatterplot, 'total_bill' , 'tip');
```

```python
tip_fg2 = sns.FacetGrid(data = tip_df, col = 'sex', hue = 'smoker',
                        height = 4, aspect = 1.5)
tip_fg2.map(sns.scatterplot, 'total_bill', 'tip',s=150, edgecolor = 'b', alpha = 0.4);
tip_fg2.add_legend(loc = 'upper right');
```
